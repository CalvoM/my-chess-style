I would like to learn about using Seaborn. I started working with Kaggle and this was one of the tutorials recommended from [Seaborn Tutorial](https://elitedatascience.com/python-seaborn-tutorial)

```python
# Normal setup and importing of libraries needed
import pandas as pd
from matplotlib import pyplot as plt

%matplotlib inline
import seaborn as sns

df = pd.read_csv("Pokemon.csv", index_col=0, encoding="latin")
```

Let us inspect the data briefly.

```python
df.head()
```

## Scatterplot

```python
sns.lmplot(data=df, x="Attack", y="Defense", fit_reg=False, hue="Legendary")
plt.ylim(0, 200)
plt.xlim(0, 160)
```

## Boxplot

```python
plt.figure(figsize=(9, 6))  # Set plot dimensions
sns.boxplot(data=df)
```
